pie chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
```

```python
data = [20, 30, 34, 45, 23, 78]
subject = ['python', 'ml', 'data science', 'full stack', 'devops', 'da']
```

```python
plt.pie(data, labels=subject)
plt.show()
```

```python
plt.pie(data, labels=subject, explode=[0,0.1,0,0,0.2,0.5], shadow= True)
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.pie(data, labels=subject, explode=[0,0.1,0,0,0.2,0.5], shadow= True, autopct='%0.2f%%')
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.pie(data, labels=subject, explode=[0,0.1,0,0,0.2,0.5], shadow= True, autopct='%0.2f%%', counterclock=False)
plt.show()
```
