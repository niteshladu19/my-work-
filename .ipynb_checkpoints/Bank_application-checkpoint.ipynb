# BANK APLLICATION

```python
from getpass import getpass
```

```python
data = {1001: ['hitesh', 'Hitesh@123', 10000, "12345"]}
```

```python
def password_validation(password):
    lower = 0
    upper = 0
    number = 0
    special = 0
    
    for i in password:
        if i.islower():
            lower += 1
        elif i.isupper():
            upper += 1
        elif i.isdigit():
            number += 1
        elif not (i.isidentifier()):
       # elif i in "~!@#$%^&*()_+<>?":
            special += 1
            
    if lower>=1 and upper>=1 and number>=1 and special>=1 and len(password)>=8:
        return True 
    else:
        return False
    
    
def signup():
    user_name = input("Enter user name: ")
    password = getpass("Enter password: ")
    security_code = getpass("Enter security code: ")
    
    if password_validation(password):
        print("Intial amount of 5000 /-  is required to actiavate your account")
        print("If you are agree enter yes(y) otherwise")
        choice = input("Enter your choice y/n: ").lower()
        if choice == 'y':
            acc_no = list(data.keys())[-1] + 1
            balance = 5000
            data[acc_no] = [user_name, password, balance, security_code]
            print("Account created successfully...")
            print("Your account number is: ", acc_no)
            
        else:
            print("Sorry we can not craete account for you because initial amount is 5000/- ")
            print("Thanks for giving your time...")
            
    else:
        print("Password is not valid...")
        
        
def login():
    acc_no = int(input("Enter Your Account Number: "))
    password = getpass("Enter Your Password: ")
    
    if acc_no in data.keys() and password == data[acc_no][1]:
        print("Login successfully...")
        
        while True:
            print("1. Credit")
            print("2. Debit")
            print("3. Change Password")
            print("4. Check Balance")
            print("5. Show Details")
            print("6. Logout")
            
            
            choice = int(input("Enter Your Choice: "))
            
            if choice == 1:
                credit_amount = float(input("Enter amount for credit: "))
                data[acc_no][2] += credit_amount
                print("Amount Credited Successfully...")
                
            elif choice == 2:
                debit_amount = float(input("Enter Amount For Debit: "))
                if (data[acc_no][2] - debit_amount) >= 5000:
                    data[acc_no][2] -= debit_amount
                    print("Amount Debited Successfully...")
                    
                else:
                    print("Insufficient Balance...")
                    
            elif choice == 3:
                old_password = getpass("Enter old password: ")
                if old_password == data[acc_no][1]:
                    new_password = getpass("Enter new password: ")
                    if password_validation(new_password):
                        data[acc_no][1] = new_password
                        print("Password change successfully...")
                    else:
                        print("incorrect password...")
                        
                else:
                    print("old password does not match..try again")
                    
            elif choice == 4:
                password = getpass("Enter password: ")
                if password == data[acc_no][1]:
                    print("Your account balance is: ",data[acc_no][2])
                    
                else:
                    print("wrong password..")
            
            elif choice == 5:
                password = getpass("Enter password: ")
                if password == data[acc_no][1]:
                    print("Account Details: ")
                    print("User Name: ",data[acc_no][0])
                    print("Account No: ", acc_no)
                    
                else:
                    print("wrong password...")
                    
            elif choice == 6:
                print("successfully logout")
                break
                
    else:
        print("wrong password or account number...")
        
def forget_password():
    acc_no = int(input("Enter account number: "))
    security_code = input("Enter your security code: ")
    
    if acc_no in data.keys() and security_code == data[acc_no][3]:
        new_password = getpass("Enter new password: ")
        if password_validation(new_password):
            data[acc_no][1] = new_password
            print("password changed successfully..")
            
        else:
            print("password is not valid...")
    
    else:
        print("account no or security code does not match")
                
while True:
    print("WELCOME TO BANK APPLICATION".center(100, "*"))
    print("1. Login")
    print("2. Signup")
    print("3. Forget Password")
    print("4. Exit")
    
    query = int(input("Enter your Query: "))
    if query == 1:
        login()
    elif query == 2:
        signup()
    elif query == 3:
        forget_password()
    elif query == 4:
        print("Exit successfuly...")
        break
    else:
        print("invalid input")
```
